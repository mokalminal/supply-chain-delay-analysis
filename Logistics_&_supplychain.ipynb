```python
# Filtering for the exact dates where both tables have data
june_data = master_df[(master_df['date'] >= '2024-06-01') & (master_df['date'] <= '2024-06-25')]
june_data.head(15)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
chart_data = june_data.groupby(['region', 'carrier'])['delay_hours_avg'].mean().reset_index()
```

```python

sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

# Generate the grouped bar chart
sns.barplot(
    data=chart_data,
    x='region',               # The main category on the horizontal axis
    y='delay_hours_avg',      # The vertical numeric height
    hue='carrier',            # The sub-category that defines the colors/groups
    palette='muted'           # Use a professional, readable color palette
)

# Polishing the labels for GitHub
plt.title('Average Carrier Delay Hours by Region (June 2024)', fontsize=14, weight='bold')
plt.xlabel('Shipment Region', fontsize=12)
plt.ylabel('Avg Delay Hours', fontsize=12)
sns.despine() # Remove the top and right border lines for a clean look

plt.show()
```
